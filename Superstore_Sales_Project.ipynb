```python
df['Postal Code'] = (df['Postal Code'].fillna(0).astype(int).astype(str))
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed', dayfirst=True)
```

```python
df = df.drop_duplicates(subset=['Order ID', 'Product Name', 'Sales'])
```

```python
df = df.copy()
df.loc[:, 'Month'] = df['Order Date'].dt.month
df.loc[:, 'Year'] = df['Order Date'].dt.year
```

```python
total_sales = df['Sales'].sum()

print("Total Sales:", total_sales)
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales
```

```python
region_profit = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_profit
```

```python
monthly_sales = df.groupby(['Year','Month'])['Sales'].sum().reset_index()
monthly_sales
```
